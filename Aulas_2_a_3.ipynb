```python
import pandas  as pd # Biblioteca usada para manipulação e análise de dados (tabelas)
import plotly.express as px # Biblioteca usada para fazer gráficos
```

```python
df_principal = pd.read_excel("/content/desafios _imersão_python.xlsx", sheet_name="principal") # DataFrame da primeira página
df_principal # Executa o DataFrame
```

```python
df_açoes = pd.read_excel("/content/desafios _imersão_python.xlsx", sheet_name="total_de_açoes") # DataFrame da segunda página
df_açoes.head() # Função usada para limitar a quantidade de linhas, sendo o padrão 5 linhas quando os parênteses estão vazios
```

```python
df_ticker = pd.read_excel("/content/desafios _imersão_python.xlsx", sheet_name="ticker") # DataFrame da terceira página
df_ticker
```

```python
df_chatgpt = pd.read_excel("/content/desafios _imersão_python.xlsx", sheet_name="chatgpt") # DataFrame da quarta página
df_chatgpt
```

```python
df_analises = pd.read_excel("/content/desafios _imersão_python.xlsx", sheet_name="analises") # DataFrame da quinta e última página (que foi usada na aula 2)
df_analises
```

```python
df_principal = df_principal[["Ativo", "Data", "Último (R$)", "Var. Dia (%)"]].copy() # Faz aparecer colunas específicas sem precisar ser na mesma ordem da planilha
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_final', 'Var. Dia (%)':'Var_dia'}).copy() # A função "rename" serve renomear colunas
df_principal
```

```python
df_principal['Var_pct'] = df_principal['Var_dia'] / 100 # Cria uma nova coluna com base em outra e dividindo os valores
df_principal['Valor_inicial'] = df_principal['Valor_final'] / (df_principal['Var_pct'] + 1) # Mesma coisa, só que com outro operador
df_principal
```

```python
df_principal = df_principal.merge(df_açoes, left_on='Ativo', right_on='Código', how='left') # Combina os DataFrames, usando 'Ativo' como chave e adicionando as colunas
# (semelhante ao PROCV do Excel, só que mais eficaz)
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtde_teorica'}).copy()
df_principal = df_principal.drop(columns=['Código']) # A função "drop" serve para dropar coluna(s)
df_principal
```

```python
df_principal['Var_rs'] = (df_principal['Valor_final'] - df_principal['Valor_inicial']) * df_principal['Qtde_teorica'] # Para criar mais outra coluna, com base em outras três
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format # Faz parar de exibir números na notação científica
```

```python
df_principal['Qtde_teorica'] = df_principal['Qtde_teorica'].astype(int) # Deixa os números da coluna como int (inteiros)
df_principal
```

```python
df_principal['Resultado'] = df_principal['Var_dia'].apply(lambda x: "Subiu" if x > 0 else ("Desceu" if x < 0 else "Estável")) # Mesmo objetivo da função IF do Excel
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da empresa'])
df_principal = df_principal.rename(columns={'Idade (anos)':'Idade'}).copy()
df_principal = df_principal.drop_duplicates() # Tira as linhas repetidas *
df_principal = df_principal.drop_duplicates().reset_index(drop=True) # Ajusta a ordem dos números do index *
df_principal.index = df_principal.index + 1 # P/ mudar a contagem do index *
# * no final dos comentários é para indicar que descobri por fora das aulas e p/ é uma abreviação de "para"
df_principal
```

```python
df_principal['Faixa_etaria'] = df_principal['Idade'].apply(lambda idade: "Entre 0-20 anos" if idade <= 20 else ("Entre 21-40 anos" if idade <= 40 else ("Entre 41-60 anos" if idade <= 60 else ("Entre 61-80 anos" if idade <= 80 else "Mais de 80 anos"))))
df_principal['Var_pct'] = df_principal['Var_pct'].round(4) # A função "round" limita a quantidade de casas decimais dos números
df_principal
```

```python
# Calculando o maior valor
maior = df_principal['Var_rs'].max()

# Calculando o menor valor
menor = df_principal['Var_rs'].min()

# Calculando a média
media = df_principal['Var_rs'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Var_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Var_rs'].mean()

# Imprimindo os resultados
print(f"Maior\tR$: {maior:,.2f}")
print(f"Menor\tR$: {menor:,.2f}")
print(f"Média\tR$: {media:,.2f}")
print(f"Média de quem subiu\tR$: {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$: {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu'] # P/ aparecer apenas as empresas que tiveram resultado positivo
df_principal_subiu
```

```python
df_principal_desceu = df_principal[df_principal['Resultado'] == 'Desceu'] # P/ aparecer apenas as empresas que tiveram resultado negativo
df_principal_desceu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Var_rs'].sum().reset_index() # "Groupby" cria um resumo dos dados por meio de uma fórmula e em ordem alfabética
# (semelhante a função "UNIQUE" do Excel, e "sum" a função "SUMIF")
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Var_rs'].sum().reset_index() # "reset_index()" é um método utilizado para redefinir o índice de um DataFrame
# Sendo que os parenteses vazios são para deixa-lo em sua forma padrão (a tabela)
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Var_rs', text='Var_rs', title='Variação R$ por Resultado') # Gera um gráfico baseado nas colunas
# Sendo "x" horizontal e "y" vertical, e "tittle" usado parar definir o título do gráfico
fig.show()
```

```python
# Etapa 1 do desafio da aula 3: Mudar a formatação dos números do gráfico de barras.

# Criando o gráfico:
fig = px.bar(df_analise_saldo, x='Resultado', y='Var_rs', text='Var_rs', title='Variação R$ por Resultado')

# Alterando a formatação dos números no eixo y:
fig.update_layout(yaxis_tickformat=".2f")  # "".2f" para duas casas decimais

# Exibindo o gráfico:
fig.show()
```

```python
# Etapa 2 do desafio: Criar o gráfico de pizza com o DataFrame df_analise_segmento.

# Criando o gráfico:
fig = px.pie(df_analise_segmento, values='Var_rs', names='Segmento', title='Distribuição do Valor por Segmento')

# Exibindo o gráfico:
fig.show()
```

```python
# Terceira e última etapa: Fazer o GroupBy da categoria de idades e gerar o gráfico de barras.

# Fazendo o Groupby:
df_empresas_por_faixa_etaria = df_principal.groupby('Faixa_etaria').size().reset_index(name='Qtde_de_empresas')
df_empresas_por_faixa_etaria
```

```python
import plotly.graph_objects as go # "plotly.graph_objects" é uma parte dessa biblioteca, sendo "go" o alias (usado como um prefixo) *
```

```python
# Parte dois da terceira etapa.

# Criando o gráfico de barras horizontais:
fig = go.Figure([go.Bar(y=df_empresas_por_faixa_etaria['Faixa_etaria'], x=df_empresas_por_faixa_etaria['Qtde_de_empresas'], orientation='h', marker_color='gold')])
# "orientation" é para definir a direção das barras, senho "h' horizontal e "v" vertical
# Já "marker_color" é para definir a cor, sendo necessário escrever em inglês

# Adicionando rótulos e título:
fig.update_layout(
    xaxis=dict(title='Número de Empresas'),
    yaxis=dict(title='Faixa Etária'),
    title='Número de Empresas por Faixa Etária'
)

# Exibindo o gráfico:
fig.show()
```
